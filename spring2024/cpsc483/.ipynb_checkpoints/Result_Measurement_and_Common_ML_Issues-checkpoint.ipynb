### <b>Measuring Result

#### <b>Choose a Measuring Method

After previous class activity, we would have different results.
<br>However, how should we compare between the results? How can we say that one result is better than another?
<br>
<br>In this case, we need to choose a metric to measure how accurate your result is.

<b>Thinking Question</b>

<br>How would you measure results of your models in last class activity?

#### <b>Common Measuring Methods

For problems highly related to difference in numeric values, MSE/MAE/MAPE/MSLE are commonly used to measure results.
<br>
<br>In classification, metric based on Confusion Matrix are commonly used.
<br>
<br>There is no fixed way to measure the result. Sometimes you will need to come up with a brand new method to measure your results.
<br>
<br>Sklearn provides a number of commonly used metrics in their API. You may check their [documentation](https://scikit-learn.org/stable/modules/classes.html#sklearn-metrics-metrics) for more details.

#### <b>Cross Validation

In most problems, we will not have data of all possible cases. Meaning that our models will need to give results on data not in the training dataset.
<br>In this case, we would care about generalization gap of our model. A common method to measure it is cross validation.
<br>
<br> Cross validation is to split data into a training set and a validation set and train your models with the training set and evaluate them with the validation set.
<br>
<br> K-fold cross-validation is a more rigid way to do cross validation. The disadvantage of such method is that it needs much more computation and thus not very applicable for complex models.
<br>
<br>You may make use of sklearn's [train_test_split](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html#sklearn.model_selection.train_test_split) function for spliting dataset.


### <b>Common ML Issues

#### <b>Overfitting

Overfiting means that the model performs well on the training data, but it does not perform on data outside training data (generalization gap is huge).
<br>
<br>Cross validation helps with this issue, but does not control it during the training process.
<br>
<br> Another way to control it is through regularization. Regularization is usually done by setting limitations on hyperparameters. <br> In the case of previous class activity, an example of regularization would be limiting the slope or intercept of the regression line.

image.png

visualization of overfitting ([source](https://en.wikipedia.org/wiki/Overfitting#/media/File:Overfitting.svg))

#### <b>Underftting

Underfitting is basically the opposite of overfitting. Your model is not complex enough to fit your data.
<br>An example would be trying to fit a 1st degree polynomial (linear) model on data with a 2nd degree polynomial like distribution.

#### <b>Low Quality Data

If your training data is full of errors, outliers, and noise, it will make it harder for the system to detect the underlying patterns, thus your system is less likely to perform well.
<br>Spending time cleaning training data is usually recommended if it is feasible.
<br>
<br>In some cases, your data might not be representing the entire population in the first place (sampling bias), and there is no simple and universal approach to alleviate it.
<br>
<br>An [interesting paper](https://arxiv.org/abs/2306.11644) you might want to take a look at on this issue.


#### <b>Data Availability

In a research conducted by Microsoft in 2001, researchers found that very different Machine Learning algorithms, including fairly simple ones, performed almost identically well on a complex problem of natural language
disambiguation once they were given enough data.
<br>
<br>The idea that data matters more than algorithms for complex problems was further popularized by Peter Norvig et al. in a paper titled “The Unreasonable Effectiveness of Data” he published. It should be noted, however, that small and medium sized datasets are still very common, and it is not always easy or cheap to get extra training data.

#### <b>Irrelevant Features

Irrelevant features might be included in raw data. They are likely to have negative effect on the result if they are not cleaned out of the training dataset.
<br>
<br>Typically, people performe <font color="lime"><b>feature selection</b></font> and <font color="lime"><b>feature extraction</b></font> on dataset before they are feed to models.
<br>
<br>Feature Selection: select only the most useful and related features for training.
<br>Feature Extraction: create additional features from existing features.
<br>
<br>In the field of data science, feature engineering takes 20% to 80% of total work time.
<br>Luckily, nowadays we have automated feature engineering tools like [Featuretools](https://featuretools.alteryx.com/en/stable/), [tsfresh](https://tsfresh.readthedocs.io/en/latest/), and [autofeat](https://pypi.org/project/autofeat/) that can save us lots of time.
