## <b>Machine Learning and Artificial Neural Networks

### <b>What is Artificial Neural Network?</b>


Artificial Neural Network (ANN) is a subset of machine learning and is the core of modern day deep learning algorithms. Its recognition began with its mathematically proven ability to approximate complex functions. However, it did not become widely popular until around the 2010s, due to the lack of computational power.
<br>
<br>ANN’s structure is inspired by the human brain. While connected biological neurons allow brain signals to go through, connected ANN neurons allow computations to go through.


##<b>Universal Approximation Theorem</b>



### <b>Approximation</b>

Why do we use approximation? Because in some cases exact solution values or expressions are extremely hard to find.
<br>
<br>[Taylor series approximation](https://en.wikipedia.org/wiki/Artificial_neural_network) is an example of such a method from your past calculus classes.
<br>
<br>However, with ANN, we are not just trying to approximate values at certain x. We are trying to approximate the entire function.



### <b>Universal Approximation Theorem in Simple English</b>

The effectiveness of feed-forward artificial neural network is backened by the <font color="lime"><b>universal approximation theorem</b></font>.
<br>
<br>In short English words, this theorem imply that neural networks can approximate a wide variety of functions with appropriate weights and non-zero errors.
<br>
<br>Such approximation has form $g(x)=C\cdot(\sigma\,\circ\,(A\cdot x+b))$

## <b>Sample Approximation with Feed-Forward Neural Network</b>

Sample approximation of function $y=x^3-50x^2-500x+1000$

#### Plot of the function:

```python
import matplotlib.pyplot as plt
import numpy as np


xc = np.linspace(-150, 150, 100000)
yc = xc**3 - 50*xc**2 - 500*xc + 1000

plt.plot(xc, yc)
plt.show()
```

#### Building the model:
<br>


```python
from tensorflow import keras
from keras import layers, activations
from keras.models import Sequential
from keras.layers import Input, Dense, Dropout, Activation, Flatten

X_train = xc
Y_train = yc

model = Sequential()
model.add(Input(shape=(1,)))
model.add(Dense(150,activation = activations.relu))
model.add(Dense(150,activation = activations.leaky_relu))
model.add(Dense(150,activation = activations.elu))
model.add(Dense(150,activation = activations.relu))
model.add(Dense(1))

model.compile(
   loss = 'MAE', optimizer = 'Adam', metrics=['mae']
)
model.fit(X_train, Y_train, batch_size = 128, epochs = 50)
```

#### Checking Result:

```python
py = model.predict(xc)

plt.plot(xc, py)
plt.plot(xc, yc)
plt.title("Model Fit")
plt.show()
es = []
for i in range(len(py)):
  es.append(yc[i] - py[i])
plt.plot(xc, es)
plt.title("Error")
plt.show()
```

### <b>How It Propagates

To reach the output layer of this neural network and get the result(output), we need to go through the following steps for all layers in between: <br> &emsp;
<font size=5>**step1**</font> Multiply layer input by weights and add all together
<br> &emsp;
<font size=5>**step2**</font> Add bias
<br> &emsp;
<font size=5>**step3**</font> Apply activation function (not required)
<br>

#### Step1 Matrix Multiplication

##### Input

Here we will use input layer as an example.
<br> Let's assign some values to our input layer.

```python
import numpy as np
x1 = 1
x2 = 2
x3 = 3

print('input layer:')
input_layer = np.array([[x1],
             [x2],
             [x3]])
print(input_layer)
print('shape:')
print(input_layer.shape)
print()
```

##### Weights

From the graph, a weight is a line that connects a neuron to another neuron.
<br> For example, the line connects $a_1^{[0]}$ (x1) and $a_1^{[1]}$ is denoted as <b>w11</b>,
<br>and line connects $a_3^{[0]}$ (x3) and $a_2^{[1]}$ is denoted as <b>w32</b>.

<br>Now lets also assign some values to <b>weights</b>.

```python
w11 = 4; w12 = 5; w13 = 6; w14 = 7
w21 = 6; w22 = 7; w23 = 8; w24 = 9
w31 = 8; w32 = 9; w33 = 1; w34 = 2

print('input layer weights:')
layer1_weights = np.array([[w11,w12,w13,w14],
              [w21,w22,w23,w24],
              [w31,w32,w33,w34]])
print(layer1_weights)
print('shape:')
print(layer1_weights.shape)
print()
```

##### Multiplication

Our aim at this step is to find:
<br> &emsp; <b> $Σw_{ij}*x_i$ </b>
<br>for each neuron at the next layer.
<br>
<br>Using Figure 1 as an example,
<br>we are trying to find
<br>$w_{11}x_1 + w_{21}x_2 + w_{31}x_3$ for $a_1^{[1]}$
<br>and
<br>$w_{12}x_1 + w_{22}x_2 + w_{32}x_3$ for $a_2^{[1]}$
<br>
<br> To get this result, we can make use of matrix multiplication.
<br> We get what we are trying to find by <font color="lime">multiplying <b>transpose</b> of matrix that contains weights with matrix that contains input</font>.
<br>
<br>Which is
<br>$layer1weights^\intercal \times input\ layer\ values$



##### Lets check if they are the same

```python
multiplication_result = np.dot(layer1_weights.T,input_layer) # T stands for transpose
print('after matrix multiplication we get matrix:')
print(multiplication_result)
print('shape:')
print(multiplication_result.shape)
print()

print('expected outputs are:')
sigma_n1 = w11*x1+w21*x2+w31*x3
print('a[1]1 = Σw11*x1+w21*x2+w31*x3 = '+str(sigma_n1))
sigma_n2 = w12*x1+w22*x2+w32*x3
print('a[1]2 = Σw12*x1+w22*x2+w32*x3 = '+str(sigma_n2))
```

#### Step2 Add Bias

Recall the form we have from universal approximation theorem
<br>$g(x)=C\cdot(\sigma\,\circ\,(A\cdot x+b))$
<br>
<br>From the multiplication, we have done the job for getting Ax part of each neuron. But we still need the $+\ b$ and $\sigma\,\circ(...)$ part.
<br>
<br>Bias is for the $+\ b$ part. At this step, we simply add bias to the the previous result.
<br>
<br>Usually, bias of a layer are all being placed into one matrix to simplify calculation steps.

####Step3 Apply Activation Function

The $\sigma\,\circ(...)$ part, [by definition](https://en.wikipedia.org/wiki/Universal_approximation_theorem) represents applying activation functions to the part enclosed by brackets.
<br>
<br>An activation function allows you to "reshape" your function into different shapes.
